# Adding an new item to the Inventory which is already present in record

```python
fd=open("record.json","r")
r=fd.read()
fd.close

record=json.loads(r)

print("\nWelcome Inventory Manager")

print("Here are the record of product :\n")
print(record)

#Add to the stock
productId = input("\nEnter Product Id : ")
name = input("Enter Name : ")
price = int(input("Enter Price : "))
quantity = int(input("Enter Quantity : "))
company = input("Enter Company name : ")
type = input("Enter type : ")

if (productId in record.keys()):
    record[productId]["qty"] += quantity
else:
    record[productId] = {"name": name ,"price": price ,"qty": quantity,"company":company,"type":type}

js=json.dumps(record)
fd=open("record.json","w")
fd.write(js)
fd.close()

print("\n")
print(record) #New record after addition of an product
```

# If the product is out of stock

```python
#purchasing from the stock

print("\nWelcome Customer")
print("Here are product list :\n")
print(record)

user_ProductId=input("\nEnter Product Id : ")
user_Quantity=int(input("Enter Quantity : "))
t=time.ctime()
if (user_ProductId in record.keys()):
    if(record[user_ProductId]["qty"] < user_Quantity):
        print(record[user_ProductId]["name"],"Out of stock !!")
    else:
        print("--------------------------------------------------------------------------------------------")
        print("Bill Generated                       Time:",t)
        print("--------------------------------------------------------------------------------------------")
        print("{:<23} {:<18} {:>16}".format("Product","Unit Rate","Amount"))
        print("____________________________________________________________________________________________")
        print("{:<25} {:<20} {:>10}".format(record[user_ProductId]["name"],record[user_ProductId]["price"],
                                            record[user_ProductId]["price"] *(user_Quantity)))
        print("____________________________________________________________________________________________")
        print("Total Amount {:>44}".format(record[user_ProductId]["price"] *(user_Quantity) ))
        print("============================================================================================")
        print("\nHave a nice day.")
        print("\n                              * * * * * * * * * *                               \n")
        record[user_ProductId]["qty"] -= user_Quantity 
    
else:
    print("\nProduct Id not found!!!\nPlease enter correct product Id.")

print("\n")
print(record)
```
